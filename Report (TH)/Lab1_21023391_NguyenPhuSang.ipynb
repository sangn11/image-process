```python
img_xam_4bit = (pic // 16) # Giảm xuống 16 mức xám
img_xam_2bit = (pic // 64) # Giảm xuống 4 mức xám
img_xam_1bit = (pic // 128)# Giảm xuống 2 mức xám

plt.figure(figsize=(15, 5))
plt.subplot(1, 4, 1)
plt.imshow(pic, cmap = 'gray')
plt.title('Ảnh xám gốc')
plt.subplot(1, 4, 2)
plt.imshow(img_xam_4bit, cmap = 'gray')
plt.title('Ảnh xám 4 bit')
plt.subplot(1, 4, 3)
plt.imshow(img_xam_2bit, cmap = 'gray') 
plt.title('Ảnh xám 2 bit')
plt.subplot(1, 4, 4)
plt.imshow(img_xam_1bit, cmap = 'gray')
plt.title('Ảnh xám 1 bit')
plt.show()
```
