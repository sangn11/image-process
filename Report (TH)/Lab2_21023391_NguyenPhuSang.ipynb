```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('goku.png')
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
grey_img = img[:,:,0]*0.07+img[:,:,1]*0.72+img[:,:,2]*0.21
```

```python
h, w = img.shape[:2]
print("height %d" % h)
print("width %d" % w)

for y in range(h // 2):
    for x in range(w):
        grey_img[h - y - 1, x] = grey_img[y, x]

plt.imshow(grey_img, cmap='gray')
plt.axis('off')
```

```python
flip_img = np.flipud(grey_img)

plt.imshow(flip_img, cmap='gray')
plt.axis('off')
```
